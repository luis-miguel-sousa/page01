```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'name': ['Ana', 'Bill', 'Calvin', 'Dea', 'Elias'],
    'fruits': ['mango|orange|pear|nectarine|banana', 'orange|peach', 'pear|mango', 'mango|apple', 'nectarine|pear|mango']
})
```

```python
pd.json_normalize(df.assign(fruits=lambda df: df['fruits'].str.split('|')).fruits)
```

```python
pd.json_normalize(df.assign(fruits=lambda df: df['fruits'].str.split('|').map(lambda x: [{'fruit': y} for y in x])).fruits).applymap(lambda x: 0 if x is None else 1)
```

```python
pd.json_normalize(df['fruits'].str.split('|'))
```

```python
df.set_index('name')['fruits'].str.get_dummies(sep="|")
```

```python
pd.DataFrame(df.set_index('name').fruits.str.split('|').explode()).assign(n=1).pivot_table(columns='fruits', index='name', fill_value=0)
```

```python
pd.json_normalize(df.fruits.map(lambda x: {y: 1 for y in x.split('|')})).set_index(df.name).fillna(0).astype(int)
```

```python
pd.json_normalize(df.fruits.map(lambda x: {y: 1 for y in x.split('|')})).set_index(df.name).fillna(0).astype(int)
```
